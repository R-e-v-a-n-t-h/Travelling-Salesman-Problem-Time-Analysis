```r
df= read.table("/content/TSPtimeData.csv",sep=',',header = TRUE)
df[1]=NULL
df
```

```r
plot(df$N,df$NaiveTSPtime,type="o",col="red",lwd=2.0, xlab = "Number of nodes", ylab="Time taken (ns)")
axis(side=1, at=3:12)
lines(df$N,df$DynamicTSPtime,type = "o",col="darkgreen",add=TRUE,lwd=2.0)
lines(df$N,df$BranchAndBoundTSPtime,type = "o",col="blue",add=TRUE,,lwd=2.0)
lines(df$N,df$BacktrackTSPtime,type = "o",col="black",add=TRUE,lwd=2.0)
legend("topleft", legend = c("Naive Approach", "Dynammic Programming","Branch and Bound","Backtracking"),
       col = c("red","darkgreen","blue","black"), 
       lwd = 4)
```

```r
#From the graph too we analyze Backtracking is the slowest method and the second slowest is naive approach.
#The other two see to not differ much. Let's confirm that with a t-test
```

```r
t.test(df$DynamicTSPtime,df$BranchAndBoundTSPtime,paired=TRUE)
```

```r
#Hence the algoorithms will differ and we should take a closer look
```

```r
plot(df$N,df$DynamicTSPtime,type = "o",col="darkgreen",lwd=2.0, xlab = "Number of nodes", ylab="Time taken (ns)")
axis(side=1, at=3:12)
lines(df$N,df$BranchAndBoundTSPtime,type = "o",col="blue",add=TRUE,lwd=2.0)

legend("topleft", legend = c("Dynammic Programming","Branch and Bound"),
       col = c("darkgreen","blue"), 
       lwd = 4)
```
