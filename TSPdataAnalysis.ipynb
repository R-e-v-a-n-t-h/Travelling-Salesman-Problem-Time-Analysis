```r
df= read.table("/content/TSPtimeData.csv",sep=',',header = TRUE)
df[1]=NULL
df
```

```r
#Without manually Analyzing the data values let's assume Naive approach is the slowest method to solve TSP. To test this hypothesis we use a t-test
```

```r
t.test(df$NaiveTSPtime, df$DynamicTSPtime,paired = TRUE)
```

```r
#1 Naive approach is slower than Dyanamic Programming
```

```r
t.test(df$NaiveTSPtime, df$BranchAndBoundTSPtime,paired = TRUE)
```

```r
#2 Naive approach is slower than Branch and Bound
```

```r
t.test(df$NaiveTSPtime, df$BacktrackTSPtime,paired = TRUE)
```

```r
#3 Naive approach is faster compared to backtracking

#So assumption was wrong. We realize that Backtracking is the slowest approach
```

```r
plot(df$N,df$NaiveTSPtime,type="o",col="red",lwd=2.0, xlab = "Number of nodes", ylab="Time taken (ns)")
axis(side=1, at=3:12)
lines(df$N,df$DynamicTSPtime,type = "o",col="darkgreen",add=TRUE,lwd=2.0)
lines(df$N,df$BranchAndBoundTSPtime,type = "o",col="blue",add=TRUE,,lwd=2.0)
lines(df$N,df$BacktrackTSPtime,type = "o",col="black",add=TRUE,lwd=2.0)
legend("topleft", legend = c("Greedy Approach", "Dynammic Programming","Branch and Bound","Backtracking"),
       col = c("red","darkgreen","blue","black"), 
       lwd = 4)
```

```r
#From the graph too we analyze Backtracking is the slowest method and the second slowest is naive approach.
#Let's take a closer look at the other two which seem to be faster
```

```r
plot(df$N,df$DynamicTSPtime,type = "o",col="darkgreen",lwd=2.0, xlab = "Number of nodes", ylab="Time taken (ns)")
axis(side=1, at=3:12)
lines(df$N,df$BranchAndBoundTSPtime,type = "o",col="blue",add=TRUE,lwd=2.0)

legend("topleft", legend = c("Dynammic Programming","Branch and Bound"),
       col = c("darkgreen","blue"), 
       lwd = 4)
```
